```python
import pandas as pd

df = pd.read_csv("realtor-data.zip.csv")

# Keep only needed columns
cols = ["status", "price", "bedrooms", "bathrooms", "city", "state", "sqft"]
df = df[cols]

# Remove rows with ANY missing values in those columns
df = df.dropna(subset=cols)

# Optional: remove zero or invalid values
df = df[(df["price"] > 0) & (df["sqft"] > 0)]

df.to_csv("cleaned_data.csv", index=False)
```

```python
df.isnull().sum()
```

```python
df = df[
    (df["price"] > 10000) &   # remove super low junk listings
    (df["sqft"] > 200) &      # remove unrealistic sizes
    (df["bedrooms"] > 0) &
    (df["bathrooms"] > 0)
]
```

```python
df["price_per_sqft"] = df["price"] / df["sqft"]
```

```python
df.to_csv("final_real_estate_clean.csv", index=False)
```

```python
import pandas as pd
import sqlite3
```

```python
df = pd.read_csv("cleaned_data.csv")
```

```python
conn = sqlite3.connect("real_estate.db")
```

```python
query = "SELECT * FROM real_estate LIMIT 5"
pd.read_sql(query, conn)
```

```python
#Most expensive states
query = """
SELECT state, ROUND(AVG(price), 2) AS avg_price
FROM real_estate
GROUP BY state
ORDER BY avg_price DESC
LIMIT 10;
"""
pd.read_sql(query, conn)
```

```python
query = """
SELECT city, ROUND(AVG(price), 2) AS avg_price
FROM real_estate
GROUP BY city
ORDER BY avg_price DESC
LIMIT 10;
"""
pd.read_sql(query, conn)
```

```python
#prices vs bedroom
query = """
SELECT bedrooms, ROUND(AVG(price), 2) AS avg_price
FROM real_estate
GROUP BY bedrooms
ORDER BY bedrooms;
"""
pd.read_sql(query, conn)
```

```python
#price vs size
query = """
SELECT 
    ROUND(AVG(sqft), 2) AS avg_sqft,
    ROUND(AVG(price), 2) AS avg_price
FROM real_estate;
"""
pd.read_sql(query, conn)
```

```python
#price per sq ft
query = """
SELECT city,
       ROUND(AVG(price / sqft), 2) AS price_per_sqft
FROM real_estate
WHERE sqft > 0
GROUP BY city
ORDER BY price_per_sqft DESC
LIMIT 10;
"""
pd.read_sql(query, conn)
```

```python
#Only “for_sale” properties
query = """
SELECT status, COUNT(*) as total
FROM real_estate
GROUP BY status;
"""
pd.read_sql(query, conn)
```
